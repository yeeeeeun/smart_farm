```python
import pandas as pd

file_path = '/content/processed_data (1).csv'

columns = ['crops_id', 'measured_date', 'stem_length', 'leaf_cnt', 'leaf_width', 'leaf_length', 'stem_thick', 'fr1_cnt','pl_value', 'ei_value', 'pi_value', 'el_value']

df = pd.read_csv(file_path, usecols=columns)

print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['fr1_cnt'] = df['fr1_cnt'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df_cleaned = df.dropna()
```

```python
print(df_cleaned.head())
```

```python
df_cleaned.isnull().sum()
```

```python

import re

df_cleaned['measured_date'] = pd.to_datetime(df_cleaned['measured_date'])

def extract_number(value):
    match = re.search(r'\d+\.?\d*', str(value))
    return float(match.group()) if match else None

df_cleaned['fr1_cnt'] = df_cleaned['fr1_cnt'].apply(extract_number)

print(df_cleaned.dtypes)
```

```python
df_cleaned.info()
```

```python

df_unique = df_cleaned.groupby(['crops_id', 'measured_date']).first().reset_index()

print("\n중복 제거 후 데이터프레임:")
print(df_unique)
```

```python
final_df = df_unique.copy()
final_df = final_df.sort_values(by=['crops_id', 'measured_date'])
```

## 다음주 길이 칼럼 생성

```python


final_df['next_stem_length'] = final_df.groupby('crops_id')['stem_length'].shift(-1)
final_df['next_leaf_cnt'] = final_df.groupby('crops_id')['leaf_cnt'].shift(-1)
final_df['next_leaf_width'] = final_df.groupby('crops_id')['leaf_width'].shift(-1)
final_df['next_leaf_length'] = final_df.groupby('crops_id')['leaf_length'].shift(-1)
final_df['next_stem_thick'] = final_df.groupby('crops_id')['stem_thick'].shift(-1)

print(final_df)
```

```python
final_df.info()
```

```python
final_df_cleaned =final_df.dropna()
```

```python
final_df_cleaned.to_csv('final_data.csv', index=False)
```

```python

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

features = ['stem_length','leaf_cnt','leaf_width','leaf_length','stem_thick','fr1_cnt','ei_value', 'pl_value', 'pi_value', 'el_value']
final_df[features] = scaler.fit_transform(final_df[features])
```

#RNN 모델

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split

df = pd.read_csv('final_data.csv')

X = df[['stem_length', 'leaf_cnt', 'leaf_width', 'leaf_length', 'stem_thick', 'fr1_cnt', 'pl_value', 'ei_value', 'pi_value', 'el_value']].values
y = df[['next_stem_length', 'next_leaf_cnt', 'next_leaf_width', 'next_leaf_length', 'next_stem_thick']].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)

model = Sequential()
model.add(LSTM(64, input_shape=(1, 10)))
model.add(Dense(5))
model.compile(optimizer='adam', loss='mean_squared_error')

model.fit(X_train.reshape(-1, 1, 10), y_train, epochs=500, batch_size=32)

y_pred = model.predict(X_test.reshape(-1, 1, 10))
```

```python

print("현재 작물의 정보를 입력해주세요.")
stem_length = float(input("줄기 길이: "))
leaf_cnt = int(input("잎 개수: "))
leaf_width = float(input("잎 너비: "))
leaf_length = float(input("잎 길이: "))
stem_thick = float(input("줄기 굵기: "))
fr1_cnt = int(input("과실 개수: "))
ei_value = float(input("양액 EC: "))
pl_value = float(input("양액 PH: "))
pi_value = float(input("배지 PH: "))
el_value = float(input("배지 EC: "))

new_input = np.array([[[ stem_length, leaf_cnt, leaf_width, leaf_length, stem_thick, fr1_cnt, pl_value, ei_value, pi_value, el_value ]]])
y_pred = model.predict(new_input)

print("\n일주일 뒤의 작물 성장 예측:")
print("- 줄기 길이: {:.2f}".format(y_pred[0][0]))
print("- 잎 개수: {:.2f}".format(y_pred[0][1]))
print("- 잎 너비: {:.2f}".format(y_pred[0][2]))
print("- 잎 길이: {:.2f}".format(y_pred[0][3]))
print("- 줄기 굵기: {:.2f}".format(y_pred[0][4]))
```
